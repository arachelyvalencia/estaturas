```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 7)
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.labelsize'] = 12

print("Procesando 'altura6.dat'...")
try:
    data_cruda = pd.read_csv('altura6.dat', header=None, names=['altura'])

    plt.figure()

    # Dibujar un histograma y una curva de densidad (KDE)

    sns.histplot(data_cruda['altura'], bins=20, kde=True, color='dodgerblue')

    # Añadir títulos y etiquetas para que la gráfica sea clara
    plt.title('Distribución de Alturas de Hombres Mexicanos (Datos Crudos)')
    plt.xlabel('Altura (metros)')
    plt.ylabel('Frecuencia (número de personas)')
    plt.grid(True, linestyle='--', alpha=0.6) # Añade una cuadrícula sutil
    plt.show() # Muestra la gráfica

except FileNotFoundError:
    print("Error: No se encontró el archivo 'altura6.dat'.")
    print("Asegúrate de haberlo subido al entorno de Colab.\n")



print("\nProcesando 'altura5.dat'...")
try:
    column_names = ['altura_central', 'frecuencia_1', 'frecuencia_2', 'frecuencia_3']
    data_agrupada = pd.read_csv('altura5.dat', comment='#', delim_whitespace=True, names=column_names)


    ancho_barra = np.diff(data_agrupada['altura_central']).mean()

    plt.figure()
    plt.bar(data_agrupada['altura_central'], data_agrupada['frecuencia_1'],
            width=ancho_barra, edgecolor='black', color='skyblue',
            label='Frecuencia 1')

    plt.title('Histograma de Alturas (Datos Agrupados)')
    plt.xlabel('Altura (metros)')
    plt.ylabel('Frecuencia')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

except FileNotFoundError:
    print("Error: No se encontró el archivo 'altura5.dat'.")
    print("Asegúrate de haberlo subido al entorno de Colab.\n")

```
